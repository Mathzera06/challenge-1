```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# Lendo a planilha1 de exportação
dados_exportacao = pd.read_excel('Compilado_exportacao_vinhos_de_mesa.xlsx', sheet_name='Planilha1')
```

```python
# Cabeçalhos
dados_exportacao.head()
```

```python
dados_exportacao.isna().sum()
```

```python
# Formatando Coluna de Valor
dados_exportacao['Valor (US$)'] = dados_exportacao['Valor (US$)'] / 1000
```

```python
# Gráfico de Barras 
grafico_valor_export_ano = dados_exportacao.sort_values('Ano ', ascending=True).plot(
  x="Ano ", y="Valor (US$)", 
  kind="line", 
  figsize=(10,5), rot=45)


# Formatação do Gráfico
grafico_valor_export_ano.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.2f}M"))
plt.xlabel('Ano', fontsize=12), plt.ylabel('Valor', fontsize=12), plt.title('Valor exportado por Ano', fontsize=14)
plt.grid(axis='y', linestyle="--", alpha=0.5)

# Rótulo de Dados

plt.show()
```
