```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
df_comercializacao = pd.read_csv(r'C:\Users\T-GAMER\dados_pos\challenge-1\dados\Comercializacao.csv', encoding='UTF-8', sep=r'[;\t]', decimal=',', thousands='.', quotechar='"', engine='python')
```

```python
df_comercializacao['Produto'] = df_comercializacao['Produto'].str.strip().str.upper()

df_filtrado = df_comercializacao[df_comercializacao['Produto'] == 'VINHO DE MESA']
```

```python
linha_produto = df_filtrado.iloc[0]
anos = df_filtrado.columns[3:]  # do 1970 at√© 2023
valores = linha_produto[3:].astype(float)
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(anos, valores, marker='o', linestyle='-', color='crimson')

anos_int = anos.astype(int)
ax.set_xticks(anos[::5])  

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
plt.title('Comercializa√ß√£o de VINHO DE MESA por Ano')
plt.xlabel('Ano')
plt.ylabel('Valor (US$)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
